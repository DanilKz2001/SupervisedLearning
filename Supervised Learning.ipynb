Так, присутствует проблема дисбаланса классов. Ее исследованию посвящен следующий раздел проекта.

Напомним, что следует удалить из `data` ненужные нам признаки: таковы `CustomerId`, `Surname`. ID клиента и фамилия - индивидуальные характеристики, для каждого отдельного клиента они почти наверняка неповторимы.

```python
data_prepared = data.drop(['CustomerId', 'Surname'], axis=1)
data_prepared
```

Продолжим подготовку признаков и преобразуем категориальные признаки `Geography` и `Gender` в численные посредством техники прямого кодирования **OHE**.

Для прямого кодирования используем из библиотеки pandas функция `pd.get_dummies()`.

Чтобы избежать дамми-ловушку используем функцию с параметром `drop_first=True`. 

```python
data_prepared = pd.get_dummies(data_prepared, drop_first=True)
data_prepared
```

Произведем масштабирование количественных признаков.Используем один из методов масштабирования — стандартизацию данных.

Для отмеченной цели в `sklearn` есть отдельная структура для стандартизации данных —  `StandardScaler`. Он находится в модуле `sklearn.preprocessing`.

Создадим объект этой структуры и настроим его на обучающих данных. Настройка — это вычисление среднего и дисперсии.


Преобразуем обучающую и валидационную выборки функцией `transform()`. 

```python
scaler = StandardScaler() 

numeric = ['CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'EstimatedSalary']

scaler = StandardScaler()
scaler.fit(data_prepared[numeric])
data_prepared[numeric] = scaler.transform(data_prepared[numeric])

data_prepared
```

Правильнее обучать StandardScaler на обучающей выборке, а трансформировать уже все. Это связано с так называемым «data leaking»: мы отделяем тестовую выборку от обучающей, чтобы модель не знала о тестовой выборке ни распределения данных, ни выбросов. Так мы можем удостовериться, что модель правильно определила зависимости и будет делать корректные прогнозы. Если обучить StandardScaler на полном датасете, то при трансформации данные маштабируются с учетом всего распределения, в том числе тестовой выборки, о которой, как мы выяснили, модели знать нельзя.

### Разбиение данных на обучающую, валидационную и тестовую выборки.

Разобьем данные `data_prepared` на обучающую, валидационную и тестовую выборки - в соотношении 3:1:1.

Для этого используем `train_test_split` из библиотеки `sklearn`.

Установим параметр `shuffle=True` для перемешивания данных, а также учтем при помощи параметра `stratify` стратификацию данных (соотношение категорий уходящих и остающихся классов клиентов).

```python
target = data_prepared['Exited']
features = data_prepared.drop(['Exited'] , axis=1)

features_train, features_valid, target_train, target_valid = train_test_split(
    features, target, train_size=0.6, random_state=12345, shuffle=True, stratify=target)

features_valid, features_test, target_valid, target_test = train_test_split(
    features_valid, target_valid, train_size=0.5, random_state=12345, shuffle=True, stratify=target_valid)

print('Размер обучающей выборки:', features_train.shape, target_train.shape)
print('Размер валидационной выборки:', features_valid.shape, target_valid.shape)
print('Размер тестовой выборки:', features_test.shape, target_test.shape)
```

Итак, данные разбиты на три выборки. Можем продолжить исследовать различные модели.

### Исследование различных моделей машинного обучения

#### Модель логистической регрессии

```python
# создадим словарь, в котором будем хранить наилучшие результаты

best_results = {
    'LogisticRegression': None,
    'DecisionTreeClassifier': None,
    'RandomForestClassifier': None
}
```

```python
# произведем оптимизацию по пороговому значению

f1_best = 0.0
threshold_best = None
precision_threshold = []
recall_threshold = []
f1_threshold = []
roc_auc_threshold = []
end_threshold = 0.8

# модель логистической регрессии
model = LogisticRegression(random_state=12345, solver='liblinear')

# обучение модели
model.fit(features_train, target_train)

# получение предсказаний модели
predicted_valid = model.predict(features_valid)

# вероятности ответов
probabilities_valid = model.predict_proba(features_valid)
# вероятности положительногь классов
probabilities_one_valid = probabilities_valid[:, 1]

# вычислим первичное значение интересующей метрики f1_score на полученных ответах (без изменнеия порогового значения)
print('Значение f1_score (без предварительного изменения порогового значения) составляет {:3f}'
      .format(f1_score(target_valid, predicted_valid)))

# пробегаемся по различным порогам (до значения порога end_threshold) 
for threshold in np.arange(0, end_threshold, 0.001):
    predicted_valid = probabilities_one_valid > threshold
    precision = precision_score(target_valid, predicted_valid)
    recall = recall_score(target_valid, predicted_valid)
    f1 = f1_score(target_valid, predicted_valid)
    
    if f1 > f1_best:
        f1_best = f1
        threshold_best = threshold
    
    roc_auc = roc_auc_score(target_valid, predicted_valid)
    
    precision_threshold.append(precision)
    recall_threshold.append(recall)
    f1_threshold.append(f1)
    roc_auc_threshold.append(roc_auc)

print("Наилучший f1-score составляет {:.3f} при пороге {:.3f}".format(f1_best, threshold_best))
print('При таком пороге recall равен {:.3f}, precision достигает {:.3f}, а roc_auc_score есть {:.3f}'
      .format(recall_threshold[int(1000* threshold_best)], precision_threshold[int(1000* threshold_best)], 
             roc_auc_threshold[int(1000* threshold_best)]))

# рисовка графиков изменения метрик (с изменением порогового значения)
plt.plot(np.arange(0, end_threshold, 0.001), f1_threshold, label='f1_score')
plt.plot(np.arange(0, end_threshold, 0.001), recall_threshold, label='recall_score')
plt.plot(np.arange(0, end_threshold, 0.001), precision_threshold, label='precision_score')
plt.plot(np.arange(0, end_threshold, 0.001), roc_auc_threshold, label='roc_auc_score')
plt.axvline(x=threshold_best, color='black')
plt.xlabel('threshold')
plt.ylabel('metrics_values')
plt.title('Динамика различных метрик при изменении порогового значения')
plt.legend()
plt.show()

best_results['LogisticRegression'] = {
    'f1': f1_best,
    'auc_roc': roc_auc_threshold[int(1000* threshold_best)],
    'recall': recall_threshold[int(1000* threshold_best)],
    'precision': precision_threshold[int(1000* threshold_best)],
    'threshold': threshold_best
}
```

**Вывод**

При помощи модели логистической регрессии удалось получить значение f1 меры, равное 0.473 (порог 0.183). При таком пороге recall = 0.714, а precision = 0.354, roc_auc есть 0.690.

```python
# произведем оптимизацию по максимальной глубине дерева и по количеству деревьев

f1_best = 0.0
depth_best = None
estimators_best = None
precision_best = None
recall_best = None
roc_auc_best = None
end_depth = 21
end_estimators = 45

# пробегаемся по различным глубинам
for depth in range(15, end_depth):
    for estimators in range(40, end_estimators):
        # модель случайного леса
        model = RandomForestClassifier(random_state=12345, max_depth=depth, n_estimators=estimators)

        # обучение модели
        model.fit(features_train, target_train)

        # получение предсказаний модели
        predicted_valid = model.predict(features_valid)

        precision = precision_score(target_valid, predicted_valid)
        recall = recall_score(target_valid, predicted_valid)
        f1 = f1_score(target_valid, predicted_valid)
        roc_auc = roc_auc_score(target_valid, predicted_valid)

        if f1 > f1_best:
            f1_best = f1
            depth_best = depth
            estimators_best = estimators
            precision_best = precision
            recall_best = recall
            roc_auc_best = roc_auc


print("Наилучший f1-score составляет {:.3f} при глубине {:.3f} и количестве деревьев {:.3f}"
      .format(f1_best, depth_best, estimators_best))
print('При таких гиперпараметрах recall равен {:.3f}, precision достигает {:.3f}, а roc_auc_score есть {:.3f}'
      .format(precision_best, recall_best, roc_auc_best))
```

### Увеличение выборки (upsampling)

Сделаем объекты редкого класса 1 не такими редкими.

Преобразование выборки производим в несколько этапов:

* Разделить обучающую выборку на отрицательные и положительные объекты;
* Скопировать несколько раз положительные объекты;
* С учётом полученных данных создать новую обучающую выборку;
* Перемешать данные.

Скопировать объекты несколько раз поможет синтаксис умножения списков в Python. 

```python
# разбиение исходной выборки выполнили ранее: имеем обучающую, валидационную и тестовую выборки

# функция увеличения выборки
def upsample(features, target, repeat):
    # отделение классов
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    # копирование положительных объектов repeat раз
    features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat)
    target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat)
    
    # перемешивание данных
    features_upsampled, target_upsampled = shuffle(
        features_upsampled, target_upsampled, random_state=12345)
    
    return features_upsampled, target_upsampled

features_upsampled, target_upsampled = upsample(features_train, target_train, 4)
```

```python
# вывод размеров новой обучающей выборки

print('Размер обучающей выборки:', features_upsampled.shape, target_upsampled.shape)
```

Каждый положительный объект повторили 4 раза для сбалансирования классов. Так, теперь в обучающей выборке 50.6% положительных объектов и 49.4% отрицательных объектов.

Так, на тестовой выборке удалось достичь значения **F1 меры, равного 0.619**, что больше критического для проекта значения 0.590.

Модель правильно классифицирует **66%** объектов *положительного класса*.
При это среди *положительных предсказаний верными оказываются* больше половины: **58%**. (кажется, в этой задаче метрика recall важнее; важно охватить как можно больше клиентов, которые могут уйти; и хорошо то, что значение метрики recall уж не меньше, чем precision).

Метрика **AUC-ROC** для модели составляет **0.79**. Модель адекватна - в сравнении со случайным классификатором. 
Она способна предсказывать уход клиента лучше, чем случайная модель.

Ислледование можно продолжить и, вероятно, улучшить модели при помощи данных маркетологов, которые сравнили стоимости привлечения нового клиента и удержания текущего.
